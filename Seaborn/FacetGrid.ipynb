```python
import numpy as np
import pandas as pd
from scipy import stats,integrate
import matplotlib.pyplot as  plt
import seaborn as sns
import matplotlib as mpl
```

```python
sns.set(style='whitegrid',color_codes=True)
np.random.seed(sum(map(ord,'categorical')))
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
g=sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

```python
g=sns.FacetGrid(tips,col='sex',hue='smoker')
g.map(plt.scatter,'total_bill','tip',alpha = .7)
g.add_legend()
```

```python
from pandas import Categorical
#ordered_days = tips.day.value_counts().index
#print ordered_days
ordered_days = Categorical(['Thur','Fri','Sat','Sun'])
#print ordered_days
g = sns.FacetGrid(tips,row='day',row_order=ordered_days,height=1.7,aspect=4)
g.map(sns.boxplot,'total_bill')
```

```python
pal = dict(Lunch = 'blue',Dinner = 'red')
g =sns.FacetGrid(tips,hue = 'time',palette=pal,size=6)
g.map(plt.scatter,'total_bill','tip',s=50,alpha=.75,linewidth = .5,edgecolor='white')
g.add_legend()
```

```python
with sns.axes_style("white"):
    g = sns.FacetGrid(tips,row='sex',col='smoker',margin_titles=True,size = 2.5)
g.set_axis_labels('Total bill(US Dollars)','Tip')
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```
